```python
import hashlib
import base64
import base58
import rsa  # Lightweight RSA library

# Generate RSA key pair (2048 bits)
(public_key, private_key) = rsa.newkeys(2048)

# Your name
name = "Baryal"

# Hash the name using SHA-256
sha256_hash = hashlib.sha256(name.encode()).hexdigest()

# Hash the SHA-256 hash using RIPEMD-160
ripemd160_hash = hashlib.new('ripemd160', hashlib.sha256(name.encode()).digest()).hexdigest()

# Hash the name using SHA3-256
sha3_hash = hashlib.sha3_256(name.encode()).hexdigest()

# Encrypt the name using RSA public key
encrypted_name = rsa.encrypt(name.encode(), public_key)
encrypted_name_b64 = base64.b64encode(encrypted_name).decode()

# Sign the name using RSA private key (PKCS#1 v1.5)
signature = rsa.sign(name.encode(), private_key, 'SHA-256')
signature_b64 = base64.b64encode(signature).decode()

# Generate a Bitcoin wallet address (simplified version)
def generate_bitcoin_address(ripemd160_hash):
    # Add version byte (0x00 for Bitcoin mainnet)
    version_byte = b'\x00'
    ripemd160_bytes = bytes.fromhex(ripemd160_hash)
    payload = version_byte + ripemd160_bytes

    # Double SHA-256 hash for checksum
    checksum = hashlib.sha256(hashlib.sha256(payload).digest()).digest()[:4]

    # Combine payload and checksum
    address_bytes = payload + checksum

    # Base58 encode the result
    bitcoin_address = base58.b58encode(address_bytes)
    return bitcoin_address.decode()

bitcoin_address = generate_bitcoin_address(ripemd160_hash)

# Output results
print("Name:", name)
print("\nPrivate Key :")
print(private_key.save_pkcs1().decode())
print("\nPublic Key :")
print(public_key.save_pkcs1().decode())
print("\nSHA-256 Hash of Name:", sha256_hash)
print("RIPEMD-160 Hash of SHA-256 Hash:", ripemd160_hash)
print("SHA3-256 Hash of Name:", sha3_hash)
print("\nEncrypted Name (Base64):", encrypted_name_b64)
print("\nDigital Signature (PKCS#1 v1.5, Base64):", signature_b64)
print("\nBitcoin Wallet Address:", bitcoin_address)
```
